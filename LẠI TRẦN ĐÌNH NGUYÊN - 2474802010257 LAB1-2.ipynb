```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import warnings
warnings.filterwarnings('ignore')

# Tải dữ liệu
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
print('Shape:', df.shape)
print('Columns:', list(df.columns))
df.head()
```

```
# Câu 1: Sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
print("=== Câu 1: Sắp xếp DH1 tăng dần ===")
df_sorted_dh1 = df.sort_values(by='DH1', ascending=True)
print(df_sorted_dh1[['STT', 'DH1']].reset_index(drop=True))
```

```
# Câu 2: Sắp xếp DH2 tăng dần theo nhóm giới tính
print("=== Câu 2: Sắp xếp DH2 tăng dần theo nhóm giới tính ===")
df_sorted_gt = df.sort_values(by=['GT', 'DH2'], ascending=[True, True])
print(df_sorted_gt[['STT', 'GT', 'DH2']].reset_index(drop=True))
```

```
# Câu 3: Pivot-table thống kê DH1 theo KT
print("=== Câu 3: Pivot-table DH1 theo KT ===")

def q1(x): return x.quantile(0.25)
def q2(x): return x.quantile(0.50)
def q3(x): return x.quantile(0.75)
q1.__name__ = 'Q1'
q2.__name__ = 'Q2'
q3.__name__ = 'Q3'

pivot1 = df.groupby('KT')['DH1'].agg(['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q2, q3])
print(pivot1)
```

```
# Câu 4: Pivot-table thống kê DH1 theo KT và KV
print("=== Câu 4: Pivot-table DH1 theo KT và KV ===")
pivot2 = df.groupby(['KT', 'KV'])['DH1'].agg(['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q2, q3])
print(pivot2)
```

```
# Câu 5: Pivot-table thống kê DH1 theo KT, KV và DT
print("=== Câu 5: Pivot-table DH1 theo KT, KV và DT ===")
pivot3 = df.groupby(['KT', 'KV', 'DT'])['DH1'].agg(['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q2, q3])
print(pivot3)
```

```
# Câu 1: Trình bày dữ liệu biến GT
print("=== Câu 1: Trình bày dữ liệu biến GT ===")

# Bảng tần số và tần suất
freq_gt = df['GT'].value_counts().reset_index()
freq_gt.columns = ['GT', 'Tần số']
freq_gt['Tần suất (%)'] = (freq_gt['Tần số'] / len(df) * 100).round(2)
print(freq_gt)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Biểu đồ cột (tần số)
axes[0].bar(freq_gt['GT'], freq_gt['Tần số'], color=['steelblue', 'salmon'], edgecolor='black')
axes[0].set_title('Biểu đồ tần số - Giới tính', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Giới tính')
axes[0].set_ylabel('Tần số')
for i, v in enumerate(freq_gt['Tần số']):
    axes[0].text(i, v + 0.3, str(v), ha='center', fontweight='bold')

# Biểu đồ tròn (tần suất)
axes[1].pie(freq_gt['Tần số'], labels=freq_gt['GT'], autopct='%1.1f%%',
            colors=['steelblue', 'salmon'], startangle=90,
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[1].set_title('Biểu đồ tần suất - Giới tính', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('P2_C1_GT.png', dpi=100, bbox_inches='tight')
plt.show()
```

```
# Câu 2: Trình bày dữ liệu US_TBM1, US_TBM2, US_TBM3
print("=== Câu 2: Trình bày dữ liệu US_TBM1, US_TBM2, US_TBM3 ===")

us_cols = ['US_TBM1', 'US_TBM2', 'US_TBM3']
colors = ['#2196F3', '#4CAF50', '#FF9800']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for i, (col, color) in enumerate(zip(us_cols, colors)):
    desc = df[col].describe()
    print(f"\n{col}:\n{desc}")
    axes[i].hist(df[col], bins=15, color=color, edgecolor='black', alpha=0.8)
    axes[i].axvline(df[col].mean(), color='red', linestyle='--', label=f'Mean={df[col].mean():.2f}')
    axes[i].axvline(df[col].median(), color='green', linestyle='-', label=f'Median={df[col].median():.2f}')
    axes[i].set_title(f'Phân phối {col}', fontsize=13, fontweight='bold')
    axes[i].set_xlabel('Giá trị')
    axes[i].set_ylabel('Tần số')
    axes[i].legend()

plt.tight_layout()
plt.savefig('P2_C2_US_TBM.png', dpi=100, bbox_inches='tight')
plt.show()
```

```
# Câu 3: Trình bày dữ liệu biến DT với các học sinh là nam
print("=== Câu 3: Trình bày DT với học sinh nam ===")

df_nam = df[df['GT'] == 'M']
freq_dt_nam = df_nam['DT'].value_counts().reset_index()
freq_dt_nam.columns = ['DT', 'Tần số']
freq_dt_nam['Tần suất (%)'] = (freq_dt_nam['Tần số'] / len(df_nam) * 100).round(2)
print(f"Số học sinh nam: {len(df_nam)}")
print(freq_dt_nam)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].bar(freq_dt_nam['DT'].astype(str), freq_dt_nam['Tần số'],
            color='steelblue', edgecolor='black')
axes[0].set_title('Tần số Dân tộc - Học sinh Nam', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Dân tộc')
axes[0].set_ylabel('Tần số')

axes[1].pie(freq_dt_nam['Tần số'], labels=freq_dt_nam['DT'].astype(str),
            autopct='%1.1f%%', startangle=90)
axes[1].set_title('Tần suất Dân tộc - Học sinh Nam', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig('P2_C3_DT_Nam.png', dpi=100, bbox_inches='tight')
plt.show()
```

```
# Câu 4: Trình bày KV với học sinh nam, dân tộc Kinh, DH1>=5.0 & DH2>=4.0 & DH3>=4.0
print("=== Câu 4: Trình bày KV - nam, Kinh, điều kiện điểm ===")

mask = (df['GT'] == 'M') & (df['DT'] == 0) & (df['DH1'] >= 5.0) & (df['DH2'] >= 4.0) & (df['DH3'] >= 4.0)
df_loc = df[mask]
print(f"Số học sinh thỏa điều kiện: {len(df_loc)}")

freq_kv = df_loc['KV'].value_counts().reset_index()
freq_kv.columns = ['KV', 'Tần số']
freq_kv['Tần suất (%)'] = (freq_kv['Tần số'] / len(df_loc) * 100).round(2)
print(freq_kv)

if len(df_loc) > 0:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    axes[0].bar(freq_kv['KV'].astype(str), freq_kv['Tần số'], color='teal', edgecolor='black')
    axes[0].set_title('Tần số KV - Nam, Kinh, điều kiện điểm', fontsize=12, fontweight='bold')
    axes[0].set_xlabel('Khu vực')
    axes[0].set_ylabel('Tần số')

    axes[1].pie(freq_kv['Tần số'], labels=freq_kv['KV'].astype(str), autopct='%1.1f%%', startangle=90)
    axes[1].set_title('Tần suất KV - Nam, Kinh, điều kiện điểm', fontsize=12, fontweight='bold')

    plt.tight_layout()
    plt.savefig('P2_C4_KV.png', dpi=100, bbox_inches='tight')
    plt.show()
else:
    print("Không có dữ liệu thỏa điều kiện!")
```

```
# Câu 5: Trình bày DH1, DH2, DH3 >= 5.0 và thuộc khu vực 2NT
print("=== Câu 5: Trình bày DH1, DH2, DH3 >= 5.0 và KV = 2NT ===")

df_2nt = df[(df['KV'] == '2NT') & (df['DH1'] >= 5.0) & (df['DH2'] >= 5.0) & (df['DH3'] >= 5.0)]
print(f"Số học sinh KV=2NT và điểm >= 5.0: {len(df_2nt)}")

dh_cols = ['DH1', 'DH2', 'DH3']
colors_dh = ['#E91E63', '#9C27B0', '#3F51B5']

if len(df_2nt) > 0:
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    for i, (col, color) in enumerate(zip(dh_cols, colors_dh)):
        print(f"\n{col} - KV 2NT:\n{df_2nt[col].describe()}")
        axes[i].hist(df_2nt[col], bins=10, color=color, edgecolor='black', alpha=0.8)
        axes[i].axvline(df_2nt[col].mean(), color='black', linestyle='--',
                        label=f'Mean={df_2nt[col].mean():.2f}')
        axes[i].set_title(f'{col} - KV 2NT', fontsize=13, fontweight='bold')
        axes[i].set_xlabel('Điểm')
        axes[i].set_ylabel('Tần số')
        axes[i].legend()
    plt.tight_layout()
    plt.savefig('P2_C5_DH_2NT.png', dpi=100, bbox_inches='tight')
    plt.show()
else:
    # Nếu không đủ điều kiện 3 môn, thử lọc từng môn
    print("Thử lọc riêng từng môn >= 5.0 và KV=2NT")
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    for i, (col, color) in enumerate(zip(dh_cols, colors_dh)):
        df_tmp = df[(df['KV'] == '2NT') & (df[col] >= 5.0)]
        print(f"\n{col}: {len(df_tmp)} học sinh")
        if len(df_tmp) > 0:
            axes[i].hist(df_tmp[col], bins=8, color=color, edgecolor='black', alpha=0.8)
            axes[i].axvline(df_tmp[col].mean(), color='black', linestyle='--',
                            label=f'Mean={df_tmp[col].mean():.2f}')
        axes[i].set_title(f'{col} >= 5.0, KV=2NT', fontsize=12, fontweight='bold')
        axes[i].set_xlabel('Điểm')
        axes[i].set_ylabel('Tần số')
        axes[i].legend()
    plt.tight_layout()
    plt.savefig('P2_C5_DH_2NT.png', dpi=100, bbox_inches='tight')
    plt.show()
```

```
# Câu 2: Tạo biến phân loại phanlopt1
print("=== Câu 2: Tạo biến phân loại phanlopt1 ===")

def phan_loai_t1(score):
    if score < 5:
        return 'Kem'
    elif score < 7:
        return 'Trungbinh'
    elif score < 8:
        return 'Kha'
    else:
        return 'Gioi'

df['phanlopt1'] = df['T1'].apply(phan_loai_t1)
print(df[['T1', 'phanlopt1']].head(10))
```

```
# Câu 3: Bảng tần số cho biến phanlopt1
print("=== Câu 3: Bảng tần số phanlopt1 ===")

freq_pl = df['phanlopt1'].value_counts().reset_index()
freq_pl.columns = ['Phân loại', 'Tần số']
freq_pl['Tần suất (%)'] = (freq_pl['Tần số'] / len(df) * 100).round(2)
print(freq_pl)
```

```
# Câu 4: Biểu đồ Multiple Line cho T1 phân loại bởi phanlopt1
print("=== Câu 4: Biểu đồ Multiple Line T1 theo phanlopt1 ===")

pl_colors = {'Kem': 'red', 'Trungbinh': 'blue', 'Kha': 'orange', 'Gioi': 'green'}

plt.figure(figsize=(14, 6))
for pl, color in pl_colors.items():
    subset = df[df['phanlopt1'] == pl]
    freq = subset['T1'].value_counts().sort_index()
    plt.plot(freq.index.astype(str), freq.values, marker='o',
             label=pl, color=color, linewidth=1.5)

plt.title('Biểu đồ đường Multiple Line - T1 theo PhanlopT1', fontsize=14, fontweight='bold')
plt.xlabel('T1')
plt.ylabel('Count')
plt.xticks(rotation=90, fontsize=7)
plt.legend(title='PhanlopT1')
plt.grid(True, alpha=0.4)
plt.tight_layout()
plt.savefig('P4_C4_T1_MultipleLine.png', dpi=100, bbox_inches='tight')
plt.show()
```

```
# Câu 1: Mô tả và khảo sát phân phối biến T1
print("=== Câu 1: Mô tả phân phối T1 ===")

from scipy import stats

print("\n--- Thống kê mô tả T1 ---")
print(df['T1'].describe())
print(f"Skewness (Độ lệch): {df['T1'].skew():.4f}")
print(f"Kurtosis (Độ nhọn): {df['T1'].kurtosis():.4f}")

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Box-plot
axes[0].boxplot(df['T1'].dropna(), patch_artist=True,
                boxprops=dict(facecolor='#2196F3', color='navy'),
                medianprops=dict(color='red', linewidth=2))
axes[0].set_title('Box-Plot T1', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Điểm T1')

q1_t1, q2_t1, q3_t1 = df['T1'].quantile([0.25, 0.5, 0.75])
print(f"\nQ1={q1_t1}, Q2 (Median)={q2_t1}, Q3={q3_t1}")
print(f"IQR={q3_t1 - q1_t1}, Min={df['T1'].min()}, Max={df['T1'].max()}")
print(f"Mean={df['T1'].mean():.4f}")

# Histogram
axes[1].hist(df['T1'], bins=15, color='#4CAF50', edgecolor='black', alpha=0.8, density=True)
xmin, xmax = df['T1'].min(), df['T1'].max()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, df['T1'].mean(), df['T1'].std())
axes[1].plot(x, p, 'r-', linewidth=2, label='Phân phối chuẩn')
axes[1].set_title('Histogram T1', fontsize=13, fontweight='bold')
axes[1].set_xlabel('T1')
axes[1].set_ylabel('Mật độ')
axes[1].legend()

# QQ-Plot
stats.probplot(df['T1'].dropna(), dist='norm', plot=axes[2])
axes[2].set_title('QQ-Plot T1 (Kiểm tra phân phối chuẩn)', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig('P5_C1_PhanPhoiT1.png', dpi=100, bbox_inches='tight')
plt.show()

# Kiểm định Shapiro-Wilk
stat, p_val = stats.shapiro(df['T1'].dropna())
print(f"\nKiểm định Shapiro-Wilk: stat={stat:.4f}, p-value={p_val:.4f}")
if p_val > 0.05:
    print("→ Dữ liệu T1 có phân phối CHUẨN (p > 0.05)")
else:
    print("→ Dữ liệu T1 KHÔNG có phân phối chuẩn (p <= 0.05)")
```

```
# Câu 3: Tương quan DH1 theo T1
print("=== Câu 3: Tương quan DH1 theo T1 ===")

cov = df[['T1', 'DH1']].cov()
corr = df[['T1', 'DH1']].corr()
print(f"Covariance:\n{cov}")
print(f"\nCorrelation:\n{corr}")
print(f"\nHệ số tương quan T1-DH1: {corr.loc['T1', 'DH1']:.4f}")

plt.figure(figsize=(8, 6))
plt.scatter(df['T1'], df['DH1'], alpha=0.7, color='#2196F3', edgecolor='navy', s=60)
m, b = np.polyfit(df['T1'].dropna(), df['DH1'].dropna(), 1)
x_line = np.linspace(df['T1'].min(), df['T1'].max(), 100)
plt.plot(x_line, m * x_line + b, 'r--', linewidth=2, label=f'y={m:.2f}x+{b:.2f}')
plt.title('Scatter Plot: DH1 theo T1', fontsize=14, fontweight='bold')
plt.xlabel('T1 (Điểm Toán lớp 10)')
plt.ylabel('DH1 (Điểm thi ĐH môn 1)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('P5_C3_Scatter_DH1_T1.png', dpi=100, bbox_inches='tight')
plt.show()
```

```
# Câu 4: Tương quan DH1 theo T1 trên từng nhóm khu vực
print("=== Câu 4: Tương quan DH1 theo T1 theo từng KV ===")

kv_list = df['KV'].unique()
colors_kv = plt.cm.Set1(np.linspace(0, 1, len(kv_list)))

plt.figure(figsize=(10, 7))
for kv, color in zip(sorted(kv_list), colors_kv):
    subset = df[df['KV'] == kv]
    plt.scatter(subset['T1'], subset['DH1'], alpha=0.7, color=color, label=str(kv), s=60)

plt.title('Scatter Plot: DH1 theo T1 - Phân nhóm KV', fontsize=14, fontweight='bold')
plt.xlabel('T1')
plt.ylabel('DH1')
plt.legend(title='Khu vực')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('P5_C4_Scatter_DH1_T1_KV.png', dpi=100, bbox_inches='tight')
plt.show()

print("\nHệ số tương quan T1-DH1 theo từng KV:")
for kv in sorted(kv_list):
    subset = df[df['KV'] == kv]
    if len(subset) > 2:
        r = subset[['T1', 'DH1']].corr().loc['T1', 'DH1']
        print(f"  KV={kv}: r={r:.4f}")
```
