#VẤN ĐỀ 1


```
import pandas as pd
import numpy as np
```

```
# Thêm header
column_names = ["Id", "Name", "Age", "Weight", "m0006", "m0612", "m1218", "f0006", "f0612", "f1218"]
```

```
# Đọc file dữ liệu
df = pd.read_csv("patient_heart_rate.csv",
                 names=column_names,
                 encoding="utf-8-sig",
                 on_bad_lines="warn")

print("Dữ liệu gốc sau khi thêm header:")
print(df.to_string())
print(f"\nKích thước: {df.shape} (rows x cols)")
```

#VẤN ĐỀ 2


```
df[["Firstname", "Lastname"]] = df["Name"].str.split(expand=True)
df = df.drop("Name", axis=1)

print("Kết quả sau khi tách cột Name:")
print(df[["Id", "Firstname", "Lastname"]].to_string())
```

#VẤN ĐỀ 3

```
for i in range(len(df)):
    x = str(df.at[i, "Weight"])
    if "lbs" in x[-3:]:
        x = x[:-3]
        float_x = float(x)
        y = round(float_x / 2.2, 1)
        df.at[i, "Weight"] = str(y) + "kgs"  # gán thẳng vào df gốc

print(df[["Id", "Firstname", "Lastname", "Weight"]].to_string())
```

#VẤN ĐỀ 4


```
print(f"Số dòng trước: {len(df)}")
df.dropna(how="all", inplace=True)
df.reset_index(drop=True, inplace=True)
print(f"Số dòng sau khi xóa dòng rỗng hoàn toàn: {len(df)}")
print(df.to_string())
```

#VẤN ĐỀ 5

```
print(f"Số dòng trước: {len(df)}")
df = df.drop_duplicates(subset=["Firstname", "Lastname", "Age", "Weight"])
df.reset_index(drop=True, inplace=True)
print(f"Số dòng sau khi xóa trùng lặp: {len(df)}")
print(df.to_string())
```

#VẤN ĐỀ 6


```
print("Trước khi xử lý:")
print(df[["Firstname", "Lastname"]].to_string())

df["Firstname"] = df["Firstname"].replace({r"[^\x00-\x7F]+": ""}, regex=True)
df["Lastname"]  = df["Lastname"].replace({r"[^\x00-\x7F]+": ""}, regex=True)

print("\nSau khi xử lý non-ASCII:")
print(df[["Firstname", "Lastname"]].to_string())
```

#VẤN ĐỀ 7

```
# Chuyển Age sang numeric
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

# Tách số khỏi Weight để tính mean
df["Weight_num"] = df["Weight"].str.replace("kgs", "", regex=False).str.strip()
df["Weight_num"] = pd.to_numeric(df["Weight_num"], errors="coerce")

# Thống kê dữ liệu thiếu
print("Thống kê dữ liệu thiếu trên từng biến:")
print(f"  Age    : {df['Age'].isna().sum()} giá trị thiếu / {len(df)} dòng")
print(f"  Weight : {df['Weight_num'].isna().sum()} giá trị thiếu / {len(df)} dòng")

age_mean    = df["Age"].mean()
weight_mean = df["Weight_num"].mean()
print(f"\nMean Age    = {age_mean:.2f}")
print(f"Mean Weight = {weight_mean:.2f} kgs")

# Xóa dòng thiếu CẢ Age lẫn Weight
mask_both = df["Age"].isna() & df["Weight_num"].isna()
print(f"\nSố dòng thiếu cả Age lẫn Weight (sẽ xóa): {mask_both.sum()}")
df = df[~mask_both].copy()
df.reset_index(drop=True, inplace=True)

# Thay Age thiếu bằng mean
df["Age"] = df["Age"].fillna(round(age_mean))

# Thay Weight thiếu bằng mean
df["Weight_num"] = df["Weight_num"].fillna(round(weight_mean, 1))
df["Weight"] = df["Weight_num"].apply(lambda x: f"{x}kgs")

print("\nSau khi xử lý Missing Values:")
print(df[["Id", "Firstname", "Lastname", "Age", "Weight"]].to_string())
```

# VẤN ĐỀ 8

```
id_vars    = ["Id", "Age", "Weight", "Firstname", "Lastname"]
value_vars = ["m0006", "m0612", "m1218", "f0006", "f0612", "f1218"]

# Melt Sex + time vào 1 cột
df_melted = pd.melt(df,
                    id_vars=id_vars,
                    value_vars=value_vars,
                    value_name="PulseRate",
                    var_name="sex_and_time")

# Tách sex_and_time → Sex, hours_lower, hours_upper
tmp_df = df_melted["sex_and_time"].str.extract(r"([mf])(\d{2})(\d{2})")
tmp_df.columns = ["Sex", "hours_lower", "hours_upper"]
tmp_df["Time"]  = tmp_df["hours_lower"] + "-" + tmp_df["hours_upper"]

# Ghép lại
df_final = pd.concat([df_melted, tmp_df[["Sex", "Time"]]], axis=1)
df_final = df_final.drop(["sex_and_time", "hours_lower", "hours_upper",
                           "Weight_num"], axis=1, errors="ignore")

# Sắp xếp theo Id
df_final = df_final.sort_values(["Id", "Sex", "Time"]).reset_index(drop=True)

print("Kết quả sau khi tách (20 dòng đầu):")
print(df_final[["Id", "Firstname", "Lastname", "Age", "Weight",
                "PulseRate", "Sex", "Time"]].head(20).to_string())
```

# VẤN ĐỀ 9

```
# Thay "-" thành NaN, rồi convert sang số
df_final["PulseRate"] = df_final["PulseRate"].replace("-", np.nan)
df_final["PulseRate"] = pd.to_numeric(df_final["PulseRate"], errors="coerce")

total   = len(df_final)
missing = df_final["PulseRate"].isna().sum()
print(f"Tổng dòng     : {total}")
print(f"Thiếu PulseRate: {missing} ({missing/total*100:.1f}%)")

# Tạo FullName để group theo từng người
df_final["FullName"] = (df_final["Firstname"].fillna("") + " " +
                        df_final["Lastname"].fillna("")).str.strip()

def fill_pulse_rate(df):
    df = df.copy()

    for name in df["FullName"].unique():
        mask       = df["FullName"] == name
        person_idx = df[mask].index.tolist()
        values     = df.loc[person_idx, "PulseRate"].tolist()

        for i, idx in enumerate(person_idx):
            if not pd.isna(values[i]):
                continue

            filled = False

            # PP1: TB liền trước + liền sau
            prev_v = values[i-1] if (i > 0 and not pd.isna(values[i-1])) else None
            next_v = values[i+1] if (i < len(values)-1 and not pd.isna(values[i+1])) else None
            if prev_v is not None and next_v is not None:
                df.at[idx, "PulseRate"] = round((prev_v + next_v) / 2, 1)
                values[i] = df.at[idx, "PulseRate"]
                filled = True

            # PP2: TB 2 giá trị liền trước
            if not filled:
                prev2 = [values[j] for j in range(max(0, i-2), i)
                         if not pd.isna(values[j])]
                if len(prev2) >= 2:
                    df.at[idx, "PulseRate"] = round(np.mean(prev2), 1)
                    values[i] = df.at[idx, "PulseRate"]
                    filled = True

            # PP3: TB 2 giá trị liền sau
            if not filled:
                nxt2 = [values[j] for j in range(i+1, min(len(values), i+3))
                        if not pd.isna(values[j])]
                if len(nxt2) >= 2:
                    df.at[idx, "PulseRate"] = round(np.mean(nxt2), 1)
                    values[i] = df.at[idx, "PulseRate"]
                    filled = True

            # PP4: TB tất cả của người đó
            if not filled:
                all_vals = [v for v in values if not pd.isna(v)]
                if all_vals:
                    df.at[idx, "PulseRate"] = round(np.mean(all_vals), 1)
                    values[i] = df.at[idx, "PulseRate"]
                    filled = True

            # PP5: TB theo nhóm giới tính
            if not filled:
                sex      = df.at[idx, "Sex"]
                sex_mean = df[df["Sex"] == sex]["PulseRate"].mean()
                if not pd.isna(sex_mean):
                    df.at[idx, "PulseRate"] = round(sex_mean, 1)
                    values[i] = df.at[idx, "PulseRate"]
                    filled = True

            # PP6: TB toàn bộ hoặc mức y học ổn định
            if not filled:
                global_mean = df["PulseRate"].mean()
                df.at[idx, "PulseRate"] = (round(global_mean, 1)
                                           if not pd.isna(global_mean) else 72)
                values[i] = df.at[idx, "PulseRate"]

    return df

df_final = fill_pulse_rate(df_final)
print(f"Dữ liệu thiếu sau khi xử lý: {df_final['PulseRate'].isna().sum()}")
```

# VẤN ĐỀ 10

```
# Bỏ dòng không có tên (Firstname rỗng)
df_final = df_final[df_final["Firstname"].notna()]
df_final = df_final[df_final["Firstname"].str.strip() != ""]

# Bỏ dòng không có PulseRate hợp lệ
df_final = df_final.dropna(subset=["PulseRate"])

# Chọn cột output
output_cols = ["Id", "Age", "Weight", "Firstname", "Lastname",
               "PulseRate", "Sex", "Time"]
df_clean = df_final[output_cols].copy()

# Reindex từ 0
df_clean.reset_index(drop=True, inplace=True)

print("Dữ liệu sạch sau khi xử lý:")
print(df_clean.to_string())
print(f"\nKích thước cuối: {df_clean.shape}")

# Lưu file
df_clean.to_csv("patient_heart_rate_clean.csv", index=False)
print("\nĐã lưu file: patient_heart_rate_clean.csv")
```
